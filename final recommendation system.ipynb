```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder

# Feature selection
numerical_features = ['temperature', 'wind_speed', 'precipitation', 'humidity', 'N', 'P', 'K']
categorical_features = ['soil_type']
target = 'crop_names'

# Separate features and target variable
X = df[numerical_features + categorical_features]
y = df[target]

# One-hot encode categorical features
encoder = OneHotEncoder()
X_encoded = encoder.fit_transform(X[categorical_features])

# Concatenate numerical features with encoded categorical features
X_processed = pd.concat([X[numerical_features], pd.DataFrame(X_encoded.toarray(), columns=encoder.get_feature_names_out(categorical_features))], axis=1)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.2, random_state=42)

# Now you can proceed with model training and evaluation using X_train, X_test, y_train, and y_test
```

## SMOTE

```python
from imblearn.under_sampling import RandomUnderSampler
from sklearn.model_selection import train_test_split

# Initialize the RandomUnderSampler
under_sampler = RandomUnderSampler()

# Resample the data
x_resampled, y_resampled = under_sampler.fit_resample(encoded_df, df[target])

# Split the resampled data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x_resampled, y_resampled, test_size=0.25, random_state=0)


```

## Random forest

```python
from sklearn.metrics import classification_report, confusion_matrix

# Example with Random Forest
from sklearn.ensemble import RandomForestClassifier

model= RandomForestClassifier()
model.fit(x_resampled, y_resampled)  # Make sure to use the resampled data here
predictions = model.predict(x_test)
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, predictions)
print("Accuracy by random forest:", accuracy)

print(classification_report(y_test, predictions))
# Optionally, plot the confusion matrix
print(confusion_matrix(y_test, predictions))
```
